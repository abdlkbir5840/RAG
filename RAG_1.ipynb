```python
import pdfplumber

# Load and extract text from the CV
pdf_path = 'EL BAHMADI Abdelkabir  - CV.pdf'
with pdfplumber.open(pdf_path) as pdf:
    cv_text = " ".join(page.extract_text() for page in pdf.pages)
print(f"Debug: Extracted CV text: {cv_text[:500]}...")
```

# **Chunking**

```python
from textwrap import wrap

chunk_size = 500
chunk_overlap = 50

chunks = [
    cv_text[i:i + chunk_size]
    for i in range(0, len(cv_text), chunk_size - chunk_overlap)
]
print(f"Debug: Total Chunks: {len(chunks)}")
print(chunks)
```

# **Embedding**

```python
from sentence_transformers import SentenceTransformer
embedding_model = SentenceTransformer("all-MiniLm-L6-v2")
embeddings = embedding_model.encode(chunks, convert_to_tensor=False)
print(f"Debug: Generated {len(embeddings)} embeddings.")
```

# **Indexing with ChromaDB**

```python
import chromadb
from chromadb.config import Settings

# Specify the persistence directory
client = chromadb.Client(Settings(
    persist_directory="./chroma_db"  # Path to persist your database
))

collection = client.get_or_create_collection("cv_chunks")

documents = [{"id": str(i), "content": chunk} for i, chunk in enumerate(chunks)]
collection.add(
    embeddings=embeddings,
    metadatas=documents,
    ids=[str(i) for i in range(len(chunks))]
)

print("Debug: Chunks and embeddings indexed in ChromaDB.")
```

# **Retrieval**

```python
query = "Quel est l'experience professionnelle du candidat ?"

query_embedding = embedding_model.encode(query, convert_to_tensor=False)

results = collection.query(
    query_embeddings=[query_embedding],
    n_results=5
)

retrieved_chunks = [
    metadata["content"]
    for metadata_list in results["metadatas"]
    for metadata in metadata_list
]

retrieved_text = "\n".join(retrieved_chunks)

print(f"Retrieved Text for Prompt:\n{retrieved_text}")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, StoppingCriteria, StoppingCriteriaList

model_name = "Qwen/Qwen2.5-0.5B-Instruct"
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
template = (
    "You are an AI assistant that answers questions based on the provided context.\n"
    "If the information is not found in the context, respond with 'I don't know' and nothing else.\n"
    f"Context:\n{retrieved_text}\n"
    f"User: {query}\n"
    "Assistant: Please answer based solely on the provided context, without additional explanations."
)

print(template)

# Prepare model inputs
model_inputs = tokenizer([template], return_tensors="pt").to(model.device)
```

```python
class StopOnToken(StoppingCriteria):
    def __init__(self, stop_token: str):
        self.stop_token = stop_token

    def __call__(self, input_ids, scores, **kwargs):
        # Check if stop_token is in the current sequence
        if input_ids[0][-1] == tokenizer.encode(self.stop_token)[0]:
            return True  # Stop generation if stop_token is found
        return False

stop_criteria = StopOnToken(stop_token="I don't know")
```

```python
# Generate a response
generated_ids = model.generate(
    **model_inputs,
    max_new_tokens=528,
    # eos_token_id=tokenizer.encode(stop_token)[0]
    stopping_criteria=StoppingCriteriaList([stop_criteria])
    )
# print(f"Debug: Generated IDs: {generated_ids}")
```

```python
# Extract relevant output
generated_ids = [
    output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
]

response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
print(response)
```
